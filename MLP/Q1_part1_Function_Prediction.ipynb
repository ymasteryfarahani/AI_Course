### Import Libraries

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor
import matplotlib.pyplot as plt
```

### Implement Functions

```python
def true_function1(x):
    return (abs(np.sin(x) + x/2))

def true_function2(x):
    return np.exp(0.5* np.sin(x)) + 0.5* np.cos(np.sqrt(x+20))

def true_function3(x):
    return (np.cos(x) * x)

def true_function4(x):
    return (np.exp(-1 * x**2))
```

### Generate training data

```python
np.random.seed(0)
X = np.random.uniform(-15, 15, (100000, 1))
y1 = true_function1(X).flatten()
y2 = true_function2(X).flatten()
y3 = true_function3(X).flatten()
y4 = true_function4(X).flatten()
```

### Split data into train and test

```python
X_train, X_test, y_train1, y_test1 = train_test_split(X, y1, test_size=0.2, random_state=42)
X_train, X_test, y_train2, y_test2 = train_test_split(X, y2, test_size=0.2, random_state=42)
X_train, X_test, y_train3, y_test3 = train_test_split(X, y3, test_size=0.2, random_state=42)
X_train, X_test, y_train4, y_test4 = train_test_split(X, y4, test_size=0.2, random_state=42)
```

### Create & Train MLP Regressor

```python
mlp1 = MLPRegressor(hidden_layer_sizes=(16,16,8,8), activation='relu', solver='adam', random_state=2, max_iter=504)
mlp1.fit(X_train, y_train1)
train_score1 = mlp1.score(X_train, y_train1)
test_score1 = mlp1.score(X_test, y_test1)
print("--------------------Func1--------------------")
print(f"Training R2 Score of function1: {np.round(train_score1 * 100,2)}%")
print(f"Testing R2 Score of function1: {np.round(test_score1 * 100,2)}%")

mlp2 = MLPRegressor(hidden_layer_sizes=(16,16,8,8), activation='relu', solver='adam', random_state=2, max_iter=504)
mlp2.fit(X_train, y_train2)
train_score2 = mlp2.score(X_train, y_train2)
test_score2 = mlp2.score(X_test, y_test2)

print("--------------------Func2--------------------")
print(f"Training R2 Score of function2: {np.round(train_score2 * 100,2)}%")
print(f"Testing R2 Score of function2: {np.round(test_score2 * 100,2)}%")

mlp3 = MLPRegressor(hidden_layer_sizes=(16,16,8,8), activation='relu', solver='adam', random_state=2, max_iter=504)
mlp3.fit(X_train, y_train3)
train_score3 = mlp3.score(X_train, y_train3)
test_score3 = mlp3.score(X_test, y_test3)

print("--------------------Func3--------------------")
print(f"Training R2 Score of function3: {np.round(train_score3 * 100,2)}%")
print(f"Testing R2 Score of function3: {np.round(test_score3 * 100,2)}%")

mlp4 = MLPRegressor(hidden_layer_sizes=(16,16,8,8), activation='relu', solver='adam', random_state=2, max_iter=504)
mlp4.fit(X_train, y_train4)
train_score4 = mlp4.score(X_train, y_train4)
test_score4 = mlp4.score(X_test, y_test4)

print("--------------------Func4--------------------")
print(f"Training R2 Score of function4: {np.round(train_score4 * 100,2)}%")
print(f"Testing R2 Score of function4: {np.round(test_score4 * 100,2)}%")
```

### Generate predictions for plotting

```python
x_values = np.linspace(-15, 15, 100000).reshape(-1, 1)
y_pred1 = mlp1.predict(x_values)
y_pred2 = mlp2.predict(x_values)
y_pred3 = mlp3.predict(x_values)
y_pred4 = mlp4.predict(x_values)
```

### Plot true function vs predictions


```python
fig, ((ax1 , ax2), (ax3 , ax4)) = plt.subplots(2, 2, figsize=(10, 10))
# fig.suptitle('True Function vs MLP Prediction')
ax1.plot(x_values, true_function1(x_values), label='True Function', color='blue')
ax1.plot(x_values, y_pred1, label='MLP Predictions', color='red')
ax2.plot(x_values, true_function2(x_values), label='True Function', color='blue')
ax2.plot(x_values, y_pred2, label='MLP Predictions', color='red')
ax3.plot(x_values, true_function3(x_values), label='True Function', color='blue')
ax3.plot(x_values, y_pred3, label='MLP Predictions', color='red')
ax4.plot(x_values, true_function4(x_values), label='True Function', color='blue')
ax4.plot(x_values, y_pred4, label='MLP Predictions', color='red')
for ax in fig.get_axes():
    ax.label_outer()
```
